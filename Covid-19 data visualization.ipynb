```python
import pandas as pd
csv_path='Covid-19 Data Analysis.csv'
df=pd.read_csv(csv_path)
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
y=df["CONFIRMED CASES"]
x=df["STATES"]
z=df["ACTIVE CASES"]
w=df["DEATHS"]
plt.scatter(x,y,color='blue')
plt.scatter(x,z,color='red')
plt.scatter(x,w,color='green')
plt.title("Data visualization of Covid-19 data")
plt.xlabel("States")
plt.ylabel("Cases")
```
